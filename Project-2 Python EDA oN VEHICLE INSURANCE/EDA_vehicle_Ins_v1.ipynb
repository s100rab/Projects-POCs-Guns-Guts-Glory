```python
cust_pol_details['annual_premium(in Rs.)'] = cust_pol_details['annual_premium(in Rs.)'].fillna(cust_pol_details['annual_premium(in Rs.)'].mean())
cust_pol_details['vintage'] = cust_pol_details['vintage'].fillna(cust_pol_details['vintage'].mean())
cust_pol_details['sales_channel_code'] = cust_pol_details['sales_channel_code'].fillna(cust_pol_details['sales_channel_code'].mode().iloc[0])
cust_pol_details['response'] = cust_pol_details['response'].fillna(cust_pol_details['response'].mode().iloc[0])
cust_pol_details.isnull().sum()
```

```python
# Remove outliers from the 'age' column using the IQR_customer_details method
Q1_customer_details = customer_details['age'].quantile(0.25)
Q3_customer_details= customer_details['age'].quantile(0.75)
IQR_customer_details = Q3_customer_details - Q1_customer_details


customer_details['age'] = customer_details['age'].mask((customer_details['age'] < (Q1_customer_details - 1.5 * IQR_customer_details)) | (customer_details['age'] > (Q3_customer_details + 1.5 * IQR_customer_details)), customer_details['age'].mean())
# customer_details_cleaned = customer_details[~((customer_details['age']
customer_details['age'] 
```

```python
# # Remove outliers from the 'annual_premium','vintage' from customer policy table column using the IQR_customer_details method
Q1_cust_pol_details = cust_pol_details['annual_premium(in Rs.)'].quantile(0.25)
Q1_cust_pol_details_vin = cust_pol_details['vintage'].quantile(0.25)
Q3_cust_pol_details = cust_pol_details['annual_premium(in Rs.)'].quantile(0.75)
Q3_cust_pol_details_vin = cust_pol_details['vintage'].quantile(0.75)
IQR_cust_pol_details = Q3_cust_pol_details - Q1_cust_pol_details
IQR_cust_pol_details_vin = Q3_cust_pol_details_vin - Q1_cust_pol_details_vin


cust_pol_details['annual_premium(in Rs.)'] = cust_pol_details['annual_premium(in Rs.)'].mask((cust_pol_details['annual_premium(in Rs.)'] < (Q1_cust_pol_details - 1.5 * IQR_cust_pol_details)) | (cust_pol_details['annual_premium(in Rs.)'] > (Q3_cust_pol_details + 1.5 * IQR_cust_pol_details)), cust_pol_details['annual_premium(in Rs.)'].mean())
cust_pol_details['vintage'] = cust_pol_details['vintage'].mask((cust_pol_details['vintage'] < (Q1_cust_pol_details_vin - 1.5 * IQR_cust_pol_details_vin)) | (cust_pol_details['vintage'] > (Q3_cust_pol_details_vin + 1.5 * IQR_cust_pol_details_vin)), cust_pol_details['vintage'].mean())
cust_pol_details['annual_premium(in Rs.)']
cust_pol_details['vintage']
```

# Merge the dataset to have a master table for our insights

```python
# Merge the DataFrames based on the 'ID' column
master_table = pd.merge(customer_details, cust_pol_details, on='customer_id', how='inner')
```

```python
master_table
```

# Conclusion driven from the Master table for business use case


* i. Gender wise average annual premium


```python
gender_avg_premium = master_table.groupby('Gender')['annual_premium(in Rs.)'].mean()
gender_avg_premium.index = gender_avg_premium.index.map({0: 'male', 1: 'female'})
print("i. Gender wise average annual premium:")
print(gender_avg_premium)
```

* ii. Age wise average annual premium


```python
age_avg_premium = master_table.groupby('age')['annual_premium(in Rs.)'].mean()
print("i. Age wise average annual premium:")
print(age_avg_premium)
```

* iii. Check if the data is balanced between genders


```python
gender_counts = master_table['Gender'].value_counts()

is_balanced = abs(gender_counts.diff().fillna(0)).max() <= 1
# Print the result
print("iii. Is the data balanced between genders?")
print("Data is balanced." if is_balanced else "Data is not balanced.")
```

* iv. Vehicle age wise average annual premium


```python
vehicle_age_avg_premium = master_table.groupby('vehicle_Age')['annual_premium(in Rs.)'].mean()
vehicle_age_avg_premium.index = vehicle_age_avg_premium.index.map({3:'> 2 years',2:'1-2 year',1:'< 1 year'})
print("iv. Vehicle age wise average annual premium:")
print(vehicle_age_avg_premium)
```

* Is there any relation between Person Age and annual premium?

```python
correlation_coefficient = master_table['age'].corr(master_table['annual_premium(in Rs.)'])


# Determine the strength of the relationship based on the correlation coefficient

if correlation_coefficient < -0.5:
    relationship = "Strong negative relationship"
elif correlation_coefficient > 0.5:
    relationship = "Strong positive relationship"
else:
    relationship = "There is no relationship"

# Print the result
print("Correlation coefficient:", correlation_coefficient)
print("Relationship between Person Age and annual premium:", relationship)
```
