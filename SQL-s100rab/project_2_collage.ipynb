```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# Define your database credentials
db_config = {
    'user': 'root',
    'password': 'Sourabh97@',
    'host': 'localhost',
    'database': 'alumni'
}

def connect_to_db():
    """Establish a database connection and return the connection object."""
    try:
        connection = mysql.connector.connect(**db_config)
        if connection.is_connected():
            return connection
    except Error as e:
        print(f"Error: {e}")
        return None

def truncate_table(table_name):
    """Truncate the specified table."""
    try:
        connection = connect_to_db()
        if connection:
            cursor = connection.cursor()
            cursor.execute(f"TRUNCATE TABLE {table_name}")
            connection.commit()
            print(f"Table {table_name} truncated successfully!")
    except Error as e:
        print(f"Error: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

def insert_into_table(table_name, data):
    """Insert data into the specified table."""
    try:
        connection = connect_to_db()
        if connection:
            cursor = connection.cursor()
            cols = ", ".join(data.columns)
            vals = ", ".join(["%s"] * len(data.columns))
            query = f"INSERT INTO {table_name} ({cols}) VALUES ({vals})"
            for row in data.itertuples(index=False):
                cursor.execute(query, tuple(row))
            connection.commit()
            print(f"Data inserted successfully into {table_name}!")
    except Error as e:
        print(f"Error: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

def update_table(table_name, set_clause, where_clause):
    """Update the specified table with the given set and where clauses."""
    try:
        connection = connect_to_db()
        if connection:
            cursor = connection.cursor()
            query = f"UPDATE {table_name} SET {set_clause} WHERE {where_clause}"
            cursor.execute(query)
            connection.commit()
            print(f"Table {table_name} updated successfully!")
    except Error as e:
        print(f"Error: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()
```

```python

def read_csv_with_encoding(csv_file_path):
    """Read a CSV file with different encodings until successful."""
    encodings = ['utf-8', 'latin1', 'ISO-8859-1']
    for encoding in encodings:
        try:
            return pd.read_csv(csv_file_path, encoding=encoding)
        except UnicodeDecodeError:
            continue
    raise UnicodeDecodeError(f"Unable to decode the file {csv_file_path} with tried encodings.")
```

```python

def process_csv_to_table(table_csv_dict):
    """Process each CSV file and load its data into the corresponding table."""
    for table, csv_file_path in table_csv_dict.items():
        df = read_csv_with_encoding(csv_file_path)

        df = df.fillna('')

        truncate_table(table)

        
        insert_into_table(table, df)
```
